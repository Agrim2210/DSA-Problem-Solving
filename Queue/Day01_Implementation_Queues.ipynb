```python
#Simple queue using list
class Queue:
    def __init__(self):
        self.q=[]
    def enqueue(self,el):
        self.q.append(el)
    def dequeue(self):
        if self.q:
            return self.q.pop(0)   
               
    def front(self):
        if self.q:
            return self.q[0]
    def is_empty(self):
        return len(self.q)==0
q=Queue()
q.enqueue(12)
q.enqueue(13)
print(q.dequeue())
print(q.front())
print(q.is_empty())

''' Pop(0) takes 0(n) time (because shifting happens)
So this is not optimal'''
```

```python
# Queue using collections.deque (Best in python)
#append->o(1),popleft->o(1)
from collections import deque
q=deque()
q.append(20)
q.append(30)
print(q.popleft())
print(q[0])
```

```python
#Implementation  of Queue using stack
class Myqueue:
    def __init__(self):
        self.s1=[]
        self.s2=[]
    def push(self,el):
        self.s1.append(el)
    def pop(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2.pop()
    def peek(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2[-1]
    def is_empty(self):
        return not self.s1 and not self.s2  
q=Myqueue()
q.push(12)
q.push(13)
q.push(15)
print(q.pop())
print(q.pop())
print(q.peek())
q.is_empty()                    

                   
```
